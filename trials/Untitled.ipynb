```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
data = pd.read_csv("diabetes_data.csv")
data.Polyuria = data.Polyuria.eq('Yes').mul(True)
data.Gender = data.Gender.eq('Female').mul(True)
data.Polydipsia = data.Polydipsia.eq('Yes').mul(True)
data['sudden weight loss'] = data['sudden weight loss'].eq('Yes').mul(True)
data.weakness = data.weakness.eq('Yes').mul(True)
data.Polyphagia = data.Polyphagia.eq('Yes').mul(True)
data['Genital thrush'] = data['Genital thrush'].eq('Yes').mul(True)
data['visual blurring'] = data['visual blurring'].eq('Yes').mul(True)
data['Itching'] = data['Itching'].eq('Yes').mul(True)
data['Irritability'] = data['Irritability'].eq('Yes').mul(True)
data['delayed healing'] = data['delayed healing'].eq('Yes').mul(True)
data['partial paresis'] = data['partial paresis'].eq('Yes').mul(True)
data['muscle stiffness'] = data['muscle stiffness'].eq('Yes').mul(True)
data['Alopecia'] = data['Alopecia'].eq('Yes').mul(True)
data['Obesity'] = data['Obesity'].eq('Yes').mul(True)
data['class'] = data['class'].eq('Positive').mul(True)
data.drop('Age', inplace=True, axis=1)
```

```python
import pandas as pd
from sklearn.datasets import make_classification
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


frequent_itemsets = apriori(data, min_support=0.3, use_colnames=True)

print(frequent_itemsets)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)

# print the association rules
print(rules)
```
